This file uploads the data to neo4j using terminal commands on a macbook

```python
import subprocess
import os
```

```python
# Stop neo4j if its running
command = "neo4j stop"
result = subprocess.run(command, shell=True, check=True, capture_output=True, text=True)

# Get path of data folder
current_directory = os.getcwd()
data_folder = os.path.join(current_directory, 'data')

# Change directory to data folder
os.chdir(data_folder)
print(f"Changed directory to {data_folder}")
```

```python
# Upload data to neo4j
command = [
    "neo4j-admin", "database", "import", "full", "--overwrite-destination", "--multiline-fields=true",
    "--nodes=PAPER=papers.csv",
    "--nodes=AUTHOR=authors.csv",
    "--nodes=KEYWORD=keywords.csv",
    "--nodes=CONFERENCE_WORKSHOP=conference_workshop.csv",
    "--nodes=JOURNAL=journals.csv",
    "--relationships=WRITTEN_BY=r_written_by.csv",
    "--relationships=CORRESPONDING_AUTHOR=r_corresponding_author.csv",
    "--relationships=PUBLISHED_IN=r_published_in.csv",
    "--relationships=CITES=r_cites.csv",
    "--relationships=HAS_KEYWORD=r_has_keyword.csv",
    "--relationships=REVIEWED_CONDUCTED_BY=r_reviewed_paper.csv"
]


# Execute the command
try:
    result = subprocess.run(command, check=True, capture_output=True, text=True)
    print("Command executed successfully:")
    print(result.stdout)
except subprocess.CalledProcessError as e:
    print("Error occurred while executing command:")
    print(e.stderr)
```
